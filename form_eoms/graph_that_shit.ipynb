```python
import jax
import jax.numpy as jnp
```

```python
def lagrangian(q, dq, t):
    jax.debug.print("q {}; dq {}; t {}", q, dq, t)
    return 0.5 * 7 * jnp.dot(q, q) - 0.5 * jnp.dot(dq, dq)
```
